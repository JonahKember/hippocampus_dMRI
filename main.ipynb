```python
import json
import preprocessing, fitting, registration, noddi

with open('config.json') as f:
    config = json.load(f)


preprocessing.transform_surfaces(config)
preprocessing.transform_volumes(config)
preprocessing.upsample_DWI(config)

preprocessing.create_mask(config)
registration.refine_mask(config)

preprocessing.create_tissue_segmentation(config)
preprocessing.create_distance_volume(config)

fitting.fit_diffusion_tensors(config)
fitting.fit_neurite_volume_fraction(config)
fitting.get_voxel_directional_diffusion(config)
fitting.fit_sphere_kurtosis(config)
fitting.fit_sphere_diffusion(config)

noddi.fit_noddi(config)
```
